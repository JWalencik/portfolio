```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date, concat, lit, lpad

df_flights_raw = spark.sql("SELECT * FROM flight_data_bronze.flights")

df_flight_list = df_flights_raw.select(
    col("flight_id"),
    to_date(concat(
        col("year"),
        lit("-"),
        lpad(col("month"), 2, '0'),
        lit("-"),
        lpad(col("day"), 2, '0'),
        ),"yyyy-MM-dd").alias("date"),
    col("airline"),
    col("flight_number"),
    col("aircraft_id"),
    col("origin_airport"),
    col("destination_airport"),
    col("distance"),
    col("diverted"),
    col("cancelled"),
    col("cancellation_reason")
)

df_flight_list.createOrReplaceTempView("flight_list")
```

```python
%sql
CREATE TABLE IF NOT EXISTS flight_data_silver.dimflightlist(
  flight_id string,
  date DATE,
  airline string,
  flight_number int,
  aircraft_id STRING,
  origin_airport STRING,
  destination_airport STRING,
  distance INT,
  diverted INT,
  cancelled INT,
  cancellation_reason STRING
);

MERGE INTO flight_data_silver.dimflightlist AS dima
USING flight_list AS a
ON dima.flight_id = a.flight_id
WHEN NOT MATCHED
THEN INSERT *
```
