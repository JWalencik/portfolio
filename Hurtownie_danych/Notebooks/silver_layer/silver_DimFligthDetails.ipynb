```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

df_flights_raw = spark.sql("SELECT * FROM flight_data_bronze.flights")

df_flight_details = df_flights_raw.select(
    col("flight_id"),
    col("scheduled_departure"),
    col("departure_time"),
    col("departure_delay"),
    col("taxi_out"),
    col("wheels_off"),
    col("scheduled_time"),
    col("elapsed_time"),
    col("air_time"),
    col("wheels_on"),
    col("taxi_in"),
    col("scheduled_arrival"),
    col("arrival_time"),
    col("arrival_delay"),
    col("air_system_delay"),
    col("security_delay"),
    col("airline_delay"),
    col("late_aircraft_delay"),
    col("weather_delay")
)
df_flight_details.createOrReplaceTempView("flight_details")
```

```python
%sql
CREATE TABLE IF NOT EXISTS flight_data_silver.dimflightdetails(
  flight_id string,
  scheduled_departure string,
  departure_time string,
  departure_delay int,
  taxi_out int,
  wheels_off STRING,
  scheduled_time INT,
  elapsed_time INT,
  air_time INT,
  wheels_on STRING,
  taxi_in INT,
  scheduled_arrival STRING,
  arrival_time STRING,
  arrival_delay INT,
  air_system_delay INT,
  security_delay INT,
  airline_delay INT,
  late_aircraft_delay INT,
  weather_delay INT
);

MERGE INTO flight_data_silver.dimflightdetails AS dima
USING flight_details AS a
ON dima.flight_id = a.flight_id
WHEN NOT MATCHED
THEN INSERT *
```
