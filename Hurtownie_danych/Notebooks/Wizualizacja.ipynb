```python
%sql

SELECT MONTH(date_key) AS month, SUM(total_delay) as total_delay_in_month
FROM flight_data_gold.factflights
GROUP BY month
ORDER BY month
```

```python
%sql
SELECT DAY(date_key) AS day, SUM(total_delay) as total_delay_in_monthsday
FROM flight_data_gold.factflights
GROUP BY day
ORDER BY day
```

```python
%sql
WITH temp AS(
  SELECT flight_id, origin_airport, destination_airport FROM flight_data_gold.factflights
)

SELECT
  dima1.airport_name as origin_airport, 
  dima2.airport_name as destination_airport,
  count(t.flight_id) as flight_count
FROM temp as t
JOIN flight_data_silver.dimairports as dima1
ON t.origin_airport = dima1.airport_id
JOIN flight_data_silver.dimairports as dima2
ON t.destination_airport = dima2.airport_id
GROUP BY dima1.airport_name,  dima2.airport_name
ORDER BY flight_count DESC
LIMIT 20
```

```python
%sql

SELECT
  count(ff.origin_airport) as fligths_departed,
  dima.airport_name,
  dima.latitude,
  dima.longitude
FROM flight_data_gold.factflights ff
JOIN flight_data_silver.dimairports dima
ON ff.origin_airport = dima.airport_id
GROUP BY   dima.airport_name, dima.latitude, dima.longitude
ORDER BY fligths_departed DESC
LIMIT 20
```

```python
%sql

SELECT
  count(ff.destination_airport) as fligths_arrived,
  dima.airport_name,
  dima.latitude,
  dima.longitude
FROM flight_data_gold.factflights ff
JOIN flight_data_silver.dimairports dima
ON ff.destination_airport = dima.airport_id
GROUP BY   dima.airport_name, dima.latitude, dima.longitude
ORDER BY fligths_arrived DESC
LIMIT 20
```
