```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *

spark.sql("CREATE DATABASE IF NOT EXISTS flight_data_bronze")
spark.sql("USE flight_data_bronze")

airline_schema = StructType([
  StructField("airline_id", StringType(), True),
  StructField("airline_name", StringType(), True)
])

df_ailine = spark.read.format("csv") \
  .option("header", "true") \
  .schema(airline_schema) \
  .load("/Volumes/workspace/flight_data_bronze/flight_data/airlines.csv")

df_ailine.write.mode("overwrite").saveAsTable("flight_data_bronze.airlines")
```
